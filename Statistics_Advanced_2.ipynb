1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample of data to infer that a certain condition is true for the entire population. It involves formulating a null hypothesis and an alternative hypothesis, collecting and analyzing data, and then making a decision about whether to reject or fail to reject the null hypothesis.

2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H0) is a statement that there is no effect, no difference, or no relationship between variables. It's the default assumption that researchers aim to test and potentially reject.
- The alternative hypothesis (Ha or H1) is a statement that contradicts the null hypothesis, suggesting that there is an effect, a difference, or a relationship between variables.

3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level, often denoted as alpha (α), is a critical probability in hypothesis testing. It represents the maximum probability of rejecting the null hypothesis when it is actually true (Type I error).
- Its role is crucial in deciding the outcome of a test: if the p-value of the test statistic is less than or equal to the chosen significance level (p ≤ α), the null hypothesis is rejected.
-  Conversely, if the p-value is greater than the significance level (p > α), we fail to reject the null hypothesis.

4. What are Type I and Type II errors? Give examples of each.
- Type I error occurs when we reject a true null hypothesis.
- It's like a 'false positive'.
- An example would be a medical test incorrectly indicating a patient has a disease when they don't.


- Type II error occurs when we fail to reject a false null hypothesis.
- This is a 'false negative'.
- An example would be a medical test failing to detect a disease in a patient who actually has it.

5. What is the difference between a Z-test and a T-test? Explain when to use each.
- The Z-test is used when the population standard deviation is known and the sample size is large (typically n > 30).
- The T-test is used when the population standard deviation is unknown and the sample size is small (typically n < 30).

```python
import numpy as np
from scipy import stats

'''Implement hypothesis testing using Z-statistics for a sample dataset in Python.
Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''

sample_data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                       50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                       50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                       50.3, 50.4, 50.0, 49.7, 50.5, 49.9])


mu_0 = 50
population_std = 1.5
alpha = 0.05

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Size: {sample_size}")
print(f"Hypothesized Population Mean (mu_0): {mu_0}")
print(f"Assumed Population Standard Deviation: {population_std}")
print(f"Significance Level (alpha): {alpha}")

z_statistic = (sample_mean - mu_0) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"\nCalculated Z-statistic: {z_statistic:.4f}")
print(f"Calculated p-value: {p_value:.4f}")

print("\n--- Interpretation ---")
if p_value < alpha:
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("This suggests there is sufficient evidence to conclude that the sample mean is significantly different from the hypothesized population mean of {mu_0}.")
else:
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print(f"This suggests there is not enough evidence to conclude that the sample mean is significantly different from the hypothesized population mean of {mu_0}.")
```

```python
# : Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mu = 50
sigma = 5
sample_size = 100

np.random.seed(42)
normal_data = np.random.normal(mu, sigma, sample_size)

print(f"Simulated Data (first 5 values): {normal_data[:5].round(2)}")
print(f"Sample Mean: {np.mean(normal_data):.2f}")
print(f"Sample Standard Deviation: {np.std(normal_data):.2f}")

confidence_level = 0.95

sample_mean = np.mean(normal_data)
standard_error = stats.sem(normal_data)

df = sample_size - 1
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

margin_of_error = t_critical * standard_error

confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"\nCalculated Sample Mean: {sample_mean:.2f}")
print(f"Standard Error of the Mean (SEM): {standard_error:.2f}")
print(f"T-critical value for {confidence_level*100}% CI: {t_critical:.3f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval_lower:.2f}, {confidence_interval_upper:.2f})")

plt.figure(figsize=(10, 6))
plt.hist(normal_data, bins=15, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Simulated Normal Distribution with 95% Confidence Interval for the Mean')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)

plt.axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')

plt.axvline(confidence_interval_lower, color='green', linestyle=':', linewidth=2, label=f'95% CI Lower ({confidence_interval_lower:.2f})')
plt.axvline(confidence_interval_upper, color='green', linestyle=':', linewidth=2, label=f'95% CI Upper ({confidence_interval_upper:.2f})')

plt.legend()
plt.show()
```

```python
'''Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram.
   Explain what the Z-scores represent in terms of standard deviations from the mean.'''

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_and_visualize_zscores(data, title="Z-scores Distribution"):
    """
    Calculates Z-scores for a given dataset, visualizes them with a histogram,
    and explains their meaning.

    Args:
        data (array-like): The input dataset.
        title (str): Title for the histogram plot.
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        print("Standard deviation is zero, cannot calculate Z-scores.")
        return

    z_scores = (data - mean) / std_dev

    print(f"Original Data (first 5 values): {data[:5].round(2)}")
    print(f"Original Mean: {mean:.2f}")
    print(f"Original Standard Deviation: {std_dev:.2f}")
    print(f"\nCalculated Z-Scores (first 5 values): {z_scores[:5].round(2)}")
    print(f"Mean of Z-Scores: {np.mean(z_scores):.2f} (should be close to 0)")
    print(f"Standard Deviation of Z-Scores: {np.std(z_scores):.2f} (should be close to 1)")

    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins=20, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
    plt.title(title)
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(axis='y', alpha=0.75)
    plt.axvline(0, color='blue', linestyle='--', linewidth=1, label='Mean (0)')
    plt.axvline(1, color='green', linestyle=':', linewidth=1, label='+1 Std Dev')
    plt.axvline(-1, color='green', linestyle=':', linewidth=1, label='-1 Std Dev')
    plt.legend()
    plt.show()

    print("\n--- Explanation of Z-Scores ---")
    print("A Z-score (also called a standard score) measures how many standard deviations")
    print("an element is from the mean. It is calculated by:")
    print("  Z = (X - μ) / σ")
    print("Where:")
    print("  X is the individual data point")
    print("  μ (mu) is the population mean")
    print("  σ (sigma) is the population standard deviation")
    print("\nKey interpretations:")
    print("  - A Z-score of 0 means the data point is identical to the mean.")
    print("  - A positive Z-score indicates the data point is above the mean.")
    print("  - A negative Z-score indicates the data point is below the mean.")
    print("  - The absolute value of the Z-score tells you how far away the data point is")
    print("    from the mean in terms of standard deviations. For example, a Z-score of")
    print("    2 means the data point is 2 standard deviations above the mean.")

# Example Usage with the previously generated normal_data
print("\n--- Using Z-score function with simulated normal data ---")
calculate_and_visualize_zscores(normal_data, title='Z-scores of Simulated Normal Distribution')

# Example Usage with the previously used sample_data
print("\n--- Using Z-score function with sample data from Z-test ---")
calculate_and_visualize_zscores(sample_data, title='Z-scores of Sample Data for Z-test')
```
